```python
import pandas as pd
import numpy as np

# Read input files
enzymes_properties_path = "input_EC_and_MW_values.csv"
kcats_path = "input_kcats.xlsx"

enzymes_properties = pd.read_csv(enzymes_properties_path)
kcats_data = pd.read_excel(kcats_path, sheet_name="Sheet1")

# Create EC_number column in kcats_data
kcats_data["EC_number"] = (
    kcats_data["EC1"].astype(str) + "." +
    kcats_data["EC2"].astype(str) + "." +
    kcats_data["EC3"].astype(str) + "." +
    kcats_data["EC4"].astype(str)
)

# Function to find the closest kcat value
def find_kcat(ec_number, kcats_data):
    print(f"Processing EC_number: {ec_number}")  # Debugging log
    # Exact match
    exact_matches = kcats_data[kcats_data["EC_number"] == ec_number]
    exact_matches = exact_matches.dropna(subset=["kcat"])  # Remove NaN kcat values
    if not exact_matches.empty:
        avg_kcat = exact_matches["kcat"].median()
        print(f"Exact match found for {ec_number}: kcat = {avg_kcat}")
        return avg_kcat  # Median if multiple matches
    
    # Progressive generalization
    components = ec_number.split(".")
    for level in range(3, 0, -1):  # Generalize from most to least specific
        partial_ec = ".".join(components[:level]) + "." + ".".join(["n"] * (4 - level))
        partial_matches = kcats_data[
            kcats_data["EC_number"].str.startswith(partial_ec.split(".n")[0])
        ]
        partial_matches = partial_matches.dropna(subset=["kcat"])  # Remove NaN kcat values
        if not partial_matches.empty:
            avg_kcat = partial_matches["kcat"].median()
            print(f"Partial match found for {partial_ec}: kcat = {avg_kcat}")
            return avg_kcat  # Median for partial matches
    
    # No match found
    print(f"No match found for {ec_number}. Assigning default value of 13.7.")
    return 13.7

# Apply the function to find kcat values
enzymes_properties["kcat"] = enzymes_properties["EC_number"].apply(
    lambda ec: find_kcat(ec, kcats_data)
)

# Save the resulting table
output_path = "enzymes_kcats_MWs_medians.csv"
enzymes_properties.to_csv(output_path, index=False)

print(f"Processed data saved to '{output_path}'")
```
